```python
import keras

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPool2D
from keras.layers import Flatten

from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
x_train = train_datagen.flow_from_directory(r"split/train",target_size=(500,500),
                                            color_mode="rgb",class_mode="categorical",batch_size=32)
```

```python
x_test = test_datagen.flow_from_directory(r"split/test",target_size=(500,500),
                                          color_mode="rgb",class_mode="categorical",batch_size=32)
```

# Modeling CNN

```python
cnn = Sequential()
```

### 1st Convolution layer + Polling layer

```python
cnn.add(Conv2D(filters= 32, kernel_size=(3,3), strides=(2,2), activation="relu", input_shape = (500,500,3)))
cnn.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
```

### 2nd Convolution layer + Polling layer

```python
cnn.add(Conv2D(filters=32, kernel_size=(3,3), strides=(2,2), activation="relu"))
cnn.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
```

### Flatten layer

```python
cnn.add(Flatten())
```

### Connected layer

```python
cnn.add(Dense(units=500,kernel_initializer="random_uniform",activation="relu"))
cnn.add(Dense(units=500,kernel_initializer="random_uniform",activation="relu"))
cnn.add(Dense(units=500,kernel_initializer="random_uniform",activation="relu"))
```

### output layer

```python
cnn.add(Dense(units=8,kernel_initializer="random_uniform",activation="softmax"))
```

### Compile

```python
cnn.compile(loss=["categorical_crossentropy"], optimizer="adam", metrics=["accuracy"])
```

### summary

```python
cnn.summary()
```

### Training

```python
cnn.fit_generator(x_train,steps_per_epoch=93,epochs=10,validation_data=x_test,validation_steps=31)
```

### Saving Model

```python
cnn.save("danceform.h5")
```

### Prediction of Danceform

```python
from keras.models import load_model
import cv2
import numpy as np
```

```python
cnnmodel = load_model("danceform.h5")
```

```python
def detect(frame):
    img=cv2.resize(frame,(500,500))
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)         #scaling to 0 to 1 range
    if(np.max(img)>1):
        img = img/255.0
    img = np.array([img])
    prediction = cnnmodel.predict(img)
    label=['bharatanatyam', 'kathak', 'kathakali', 'kuchipudi', 'manipuri', 'mohiniyattam','odissi', 'sattriya']
    prediction = np.argmax(prediction)
    print("Prediction:",prediction)
    return plt.imshow(img[0]),print(label[prediction])
```
